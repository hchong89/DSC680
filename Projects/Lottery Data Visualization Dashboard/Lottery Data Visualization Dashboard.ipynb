```python
import pandas as pd
```

```python
df_powerball = pd.read_csv("Lottery_Powerball_Winning_Numbers__Beginning_2010_20240421.csv")
df_pick10 = pd.read_csv("Lottery_Pick_10_Winning_Numbers__Beginning_1987_20240421.csv")
df_mega = pd.read_csv("Lottery_Mega_Millions_Winning_Numbers__Beginning_2002_20240421.csv")
```

After loading the data, I changed draw date column to datetime 

```python
df_powerball['Draw Date'] = pd.to_datetime(df_powerball['Draw Date'])
df_pick10['Draw Date'] = pd.to_datetime(df_pick10['Draw Date'])
df_mega['Draw Date'] = pd.to_datetime(df_mega['Draw Date'])
```

```python
df_powerball['Year'] = df_powerball['Draw Date'].dt.year
df_pick10['Year'] = df_pick10['Draw Date'].dt.year
df_mega['Year'] = df_mega['Draw Date'].dt.year
```

```python
def expand_winning_numbers(df, number_col, draw_col, extra_cols=None, game_name=""):
    df = df.copy()
    df[number_col] = df[number_col].astype(str)
    df['Numbers'] = df[number_col].str.split()
    df_expanded = df[[draw_col, 'Numbers'] + (extra_cols or [])]
    df_long = df_expanded.explode('Numbers').reset_index(drop=True)
    df_long['Numbers'] = df_long['Numbers'].astype(int)
    df_long['Game'] = game_name
    return df_long
```

```python
df_powerball_clean = expand_winning_numbers(
    df_powerball,
    number_col='Winning Numbers',
    draw_col='Draw Date',
    extra_cols=['Multiplier', 'Year'],
    game_name='Powerball')
```

```python
df_pick10_clean = expand_winning_numbers(
    df_pick10,
    number_col='Winning Numbers',
    draw_col='Draw Date',
    extra_cols=['Year'],
    game_name='Pick 10'
)
```

```python
df_mega_clean = expand_winning_numbers(
    df_mega,
    number_col='Winning Numbers',
    draw_col='Draw Date',
    extra_cols=['Mega Ball', 'Multiplier', 'Year'],
    game_name='Mega Millions'
)
```

```python
df_all_cleaned = pd.concat([df_powerball_clean, df_mega_clean, df_pick10_clean], ignore_index=True)
```

```python
df_all_cleaned.to_excel("Cleaned_Lottery_Winning_Numbers.xlsx", index=False)
```
